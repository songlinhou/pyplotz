# Pyplotz

Using matplotlib with Chinese characters can be kindly of tricky. So this light weight library is designed to help you handle that problem. It simply works.

It has been tested on

* Mac OS
* Windows

More platforms will be accomendated SOON.

This is a small demo to show you how to use pyplotz.

```python
from pyplotz.pyplotz import PyplotZ
from pyplotz.pyplotz import plt
import numpy as np
```

```python
%matplotlib inline
```

```python
pltz = PyplotZ() # create an instance
```

```python
pltz.enable_chinese() # enable chinese so you can use chinese characters
```

## Work with matplotlib

```python
pltz.set_figure_size(10,5) # 
pltz.plot(np.linspace(-10,10,200),np.sin(np.linspace(-10,10,200)),'b', label='sin',alpha=0.7)
pltz.plot(np.linspace(-10,10,200),np.cos(np.linspace(-10,10,200)),'r', label='cos',alpha=0.7)
pltz.title("") # when using pltz API, you can use either "" or u"" here.
pltz.xlabel("")
pltz.ylabel("")
plt.grid() # you can use matplotlib API and pyplotz interchangeably.
pltz.legend() # use legend function from pltz to show Chinese properly
pltz.automate_font_size(scale=0.8) # always put this statement at last(scale controls the overall font size)
```

## Work with Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel('example data/example_df.xlsx')
```

```python
df.head()
```

```python
# display the entire dataframe
df.plot(figsize=(10,8))
pltz.enable_chinese()
pltz.xlabel("")
pltz.ylabel("")
pltz.legend()
```

```python
# display one column using pandas API
df[u'1'].plot(kind='bar')
pltz.xlabel("")
pltz.ylabel("1")
pltz.title("1")
pltz.legend() # this statement can help fix the chinese display using pandas
plt.xticks(rotation=360) #rotate x label using matplotlib
```

## Work with Seaborn

```python
import seaborn
import pandas as pd

pltz.enable_chinese()
seaborn.barplot(df.index.tolist(),df[u'1'],label=u"1")
pltz.xlabel("")
pltz.ylabel("1")
pltz.title("1")
pltz.legend()
pltz.title_size(18)
```
